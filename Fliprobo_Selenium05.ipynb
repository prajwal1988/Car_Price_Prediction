### OLX used cars list

```python
#getting the url
url = "https://www.olx.in/cars_c84"
driver.get(url)
time.sleep(2)
```

```python
start=0
end=5
for page in range(start,end):
    car_urls = []
    url=driver.find_elements_by_xpath('//li[@class="EIR5N"]/a')       
    for i in url:
        car_urls.append(i.get_attribute("href"))                       
car_urls
```

```python
Price=[]
try:
    price=driver.find_elements_by_xpath('//span[@class="_89yzn"]')
    for i in price:
        Price.append(i.text)
except NoSuchElementException:
    Price.append('-')
Price
```

```python
brand=[]
try:
    Brand=driver.find_elements_by_xpath('//span[@class="_2tW1I"]')
    for i in Brand:
        brand.append(i.text.split(','))
except NoSuchElementException:
    brand.append('-')
brand
```

```python
place=[]
try:
    Place=driver.find_elements_by_xpath('//span[@class="tjgMj"]')
    for i in Place:
        place.append(i.text.split(','))
except NoSuchElementException:
    place.append('-')
place
```

```python
kilometers=[]
for i in car_urls:
    driver.get(i)
    time.sleep(2)
    try:
        kms=driver.find_elements_by_xpath('//*[@id="container"]/main/div/div/div/div[5]/div[1]/div[1]/div[1]/div/div[4]/div[2]/div')
        for i in kms:
            kilometers.append(i.text)
    except NoSuchElementException:
        kilometers.append('-')
kilometers
```

```python
owners=[]
for i in car_urls:
    driver.get(i)
    time.sleep(2)
    try:
        no=driver.find_elements_by_xpath('//*[@id="container"]/main/div/div/div/div[5]/div[1]/div[2]/div/div[2]/div/div[1]/div[2]/div[2]')
        for i in no:
            owners.append(i.text)
    except NoSuchElementException:
        owners.append('-')
owners
```

```python
# creating a pandas dataframe
df = pd.DataFrame(brand, 
                  columns = ['model', 'model-year', 
                             'fuel-type','none value'])
  
# show the dataframe
df
```

```python
df.drop(columns=['none value'],axis=1,inplace=True)
```

```python
# creating a pandas dataframe
df1 = pd.DataFrame(place, 
                  columns = ['area', 'location'])
  
# show the dataframe
df1
```

```python
df1.drop(columns=['area'],axis=1,inplace=True)
```

```python
len(kilometers),len(owners),len(Price)
```

```python
olx=pd.DataFrame({'website':'olx','driven kilometers':kilometers[:530],'No of owners':owners[:530],'Price':Price[:530]})
olx
```
